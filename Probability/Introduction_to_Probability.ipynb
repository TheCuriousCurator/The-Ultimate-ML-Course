- Generate $U \sim Uniform(0,1)$
- CDF of the ideal distribution $F_X (x) = \Phi ( \frac{x - \mu}{\sigma} )$
- Therefore the transformation g is $g(U) = F^{-1}_X (U) = \sigma \Phi^{-1} (U) + \mu $


```python
# Python code to generate Gaussian from uniform
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
mu = 3
sigma = 2
U  = stats.uniform.rvs(0,1,size=100000)
gU = sigma*stats.norm.ppf(U)+mu
plt.hist(U,  bins=100); plt.show()
plt.hist(gU, bins=100); plt.show()
```

- Generate $U \sim Uniform(0,1)$
- CDF of the ideal distribution $F_X (x) = 1 - \exp(-\lambda X)$
- Therefore the transformation g is $g(U) = F^{-1}_X (U) = - \frac{1}{\lambda} \log ( 1 - U ) $


```python
# Python code to generate exponential random numbers
import numpy as np
import scipy.stats as stats

lambd = 1

U     = stats.uniform.rvs(0,1,size=100000)
gU    = -(1/lambd)*np.log(1-U)

plt.hist(U,  bins=100); plt.show()
plt.hist(gU, bins=100); plt.show()
```

# Sample Statistic

```python
# Python code to illustrate the weak law of large numbers
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy.matlib
p = 0.5
Nset = np.round(np.logspace(2,5,100)).astype(int)
x = np.zeros((1000,Nset.size))
for i in range(Nset.size):
    N = Nset[i]
    x[:,i] = stats.binom.rvs(N, p, size=1000)/N
Nset_grid = np.matlib.repmat(Nset, 1000, 1)
# each point is a sample average
plt.semilogx(Nset_grid, x, 'ko')
plt.semilogx(Nset, p + 3*np.sqrt((p*(1-p))/Nset), 'r', linewidth=6)
plt.semilogx(Nset, p - 3*np.sqrt((p*(1-p))/Nset), 'r', linewidth=6)
```
