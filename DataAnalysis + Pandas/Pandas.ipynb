# joins / merge

# mutating joins

# inner join
returns all rows from x where there are matching values in y and all columns from x and y.
If there are multiple matches between x and y, all combnations of the matches are returned.
# left join
returns all rows from x and all columns from x and y. 
Rows in x with no match in y will have NA values in the new columns. 
If there are multiple matches between x and y, all combnations of the matches are returned.
# right join
return all rows from y and all columns from x and y. 
Rows in y with no match in x will have NA values in the new column.
If there are multiple matches between x and y, all combnations of the matches are returned.
# outer join
returns all rows and all columns from both x and y. 
Where there are not matching values, returns NA for the one missing. 

# Filtering joins

# semijoin
returns all rows from x where there are matching values in y, just keeping columns from x.
inner join returns one row of x for each matching row of y, where as semijoin wil never duplicate rows of x. 
# antijoin
returns all rows from x where there are NOT matching values in y, just keeping columns from x.

```python
table1 = pd.DataFrame({"P_ID" : (1,2,3,4,5,6,7,8),
                     "gender" : ("male", "male", "female","female",
                                "female", "male", "female", "male"),
                     "height" : (71,73,64,64,66,69,62,72),
                     "weight" : (175,225,130,125,165,160,115,250)})

table1
```

```python
table2 = pd.DataFrame({"P_ID" : (1, 2, 4, 5, 7, 8, 9, 10),
                     "sex" : ("male", "male", "female","female",
                            "female", "male", "male", "female"),
                     "visits" : (1,2,4,12,2,2,1,1),
                     "checkup" : (1,1,1,1,1,1,0,0),
                     "follow_up" : (0,0,1,2,0,0,0,0),
                     "illness" : (0,0,2,7,1,1,0,0),
                     "surgery" : (0,0,0,2,0,0,0,0),
                     "ER" : ( 0,1,0,0,0,0,1,1) } ) 

table2
```

```python
out = pd.merge(table1, table2, how="inner", on="P_ID")
out
```

```python
out = pd.merge(table1, table2, how="left", on="P_ID")
out
```

```python
out = pd.merge(table1, table2, how="right", on="P_ID")
out
```

```python
out = pd.merge(table1, table2, how="outer", on="P_ID")
out
```

```python
table1.rename(columns={"gender": "sex"}, inplace=True)
out = pd.merge(table1, table2, how="outer", on=["P_ID", "sex"])
out
```

```python
review_points_mean = reviews.points.mean()
reviews.points.map(lambda p: p - review_points_mean)
```

```python
def remean_points(row):
    row.points = row.points - review_points_mean
    return row

reviews.apply(remean_points, axis="columns")
```

```python
df_grouped = reviews.groupby("country")

for group_name, df in df_grouped:
    s = 0 
    c = 0 
    for row in df["price"]:
        if pd.notnull(row):
            s = s + row
            c = c + 1
    if c==0:
        print(group_name, len(df), 0)
    else:
        print(group_name, len(df), s/c)
```

```python
df_grouped = reviews.groupby("country")

for group_name, df in df_grouped:
    s = 0 
    c = 0 
    for idx, row in df.iterrows():
        price = row.price
        if pd.notnull(price):
            s = s + price
            c = c + 1
    if c==0:
        print(group_name, len(df), 0)
    else:
        print(group_name, len(df), s/c)
```
